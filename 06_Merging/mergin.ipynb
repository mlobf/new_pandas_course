```python
# Merge horizontally
import pandas as pd
import numpy as np
```

```python
# Merge horizontally

students = pd.read_csv("students.csv")
teachers = pd.read_csv("teachers.csv")
grades1 = pd.read_csv("grades1.csv")
grades2 = pd.read_csv("grades2.csv")
contacts = pd.read_csv("contact.csv")
```

```python
# Merge horizontally

grades1
```

```python
# Merge horizontally
grades1['Semester'] = 1
```

```python
grades2['Semester'] = 2
```

```python
# Too concat two or more dataframes in a new one, use concact
grades = pd.concat([grades1, grades2])
```

```python
grades
```

```python
grades = grades.reset_index(drop=True)
grades
```

```python
# Merge horizontally
```

```python
# Merge horizontally
student_grades = pd.merge(
    students, grades, left_on="id", right_on="student_id")

student_grades
```

```python
# Merge horizontally
```

```python
student2 = students.rename({"id": "student_id"}, axis=1)
student2
```

```python
students_full = student2.merge(contacts, on='student_id')
students_full.head()
```

```python
student_grades = pd.merge(student2, grades2, on="student_id")
```

```python
student_grades
```

```python
df = pd.merge(student_grades, contacts, on="student_id")
```

```python
df = df.merge(teachers, on='course')
```

```python
df.head()
```

```python
df2 = df.loc[df.grade == "F", "student_id"]
method1 = df2.value_counts().rename_axis(
    'student_id').reset_index(name="counts")

method1
```

```python
method2 = df.loc[df.grade == "F", [
    'student_id', 'grade']].groupby("student_id").count().reset_index()
method2.columns = ['student_id', 'counts']
method2
```
